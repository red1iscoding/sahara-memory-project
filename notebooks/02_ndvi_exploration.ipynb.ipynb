```python
import ee
ee.Initialize(project="sahara-memory-project")
```

```python
algeria = ee.FeatureCollection('FAO/GAUL/2015/level0')\
            .filter(ee.Filter.eq('ADM0_NAME', 'Algeria'))
```

```python
dataset = ee.ImageCollection("MODIS/061/MOD13Q1")\
            .filterDate('2000-01-01', '2023-12-31')\
            .filterBounds(algeria)
```

```python
print(dataset.size().getInfo())
```

```python
ndvi_collection = dataset.select("NDVI")
```

```python
ndvi_collection = ndvi_collection.map(lambda image: image.divide(10000))
```

```python
mean_ndvi = ndvi_collection.mean()
```

```python
import geemap
Map = geemap.Map()
```

```python
Map.setCenter(2, 28,5)
```

```python
viz = {
    'min': 0,
    'max': 0.8,
    'palette': ['red', 'yellow', 'green']
}
```

```python
Map.addLayer(mean_ndvi, viz, 'Mean NDVI 2000-2023')
```
